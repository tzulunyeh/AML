### <font color="ce33ff">DESCRIPTION</font>:
This notebook allows you for performing data augmentation and saving the augmented images back to disc. 

-------------------------------------------------------------------------------------------------------------

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import os
```

```python
# Modified from professor's code:
# 1. Changed fill_mode='reflect' instead of default 'nearest' (avoids pixel stretching artifacts after rotation)
# 2. Removed width_shift_range and height_shift_range (prevents face cropping)
datagen = ImageDataGenerator(
    rotation_range=90,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='reflect')

# Augment ME photos
img_folder = "./data/me_cropped"
output_folder = "./data/me_aug"
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(img_folder):
    if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue
    
    img = load_img(os.path.join(img_folder, filename))
    x = img_to_array(img).reshape((1,) + img_to_array(img).shape)
    
    i = 0
    for batch in datagen.flow(x, batch_size=1, save_to_dir=output_folder,
                              save_prefix=os.path.splitext(filename)[0] + '_aug',
                              save_format='jpg'):
        i += 1
        if i >= 50:  # Fixed: generate exactly 50 images (not 51)
            break

print(f"ME: {output_folder}")

# Augment WALTER photos
img_folder = "./data/walter_cropped"
output_folder = "./data/walter_aug"
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(img_folder):
    if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue
    
    img = load_img(os.path.join(img_folder, filename))
    x = img_to_array(img).reshape((1,) + img_to_array(img).shape)
    
    i = 0
    for batch in datagen.flow(x, batch_size=1, save_to_dir=output_folder,
                              save_prefix=os.path.splitext(filename)[0] + '_aug',
                              save_format='jpg'):
        i += 1
        if i >= 50:  # Fixed: generate exactly 50 images (not 51)
            break

print(f"WALTER: {output_folder}")
print(f"Total: 3x50x2 = 300 images")
```
