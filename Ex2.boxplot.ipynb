# Task 4: Create Boxplots for Exercise 1 Datasets

## Dataset 1: Heights (Normal Distribution)

```python
# Import necessary libraries
import pandas as pd
import kagglehub

# Download and load Dataset 1: Heights and Weights
path = kagglehub.dataset_download("burnoutminer/heights-and-weights-dataset")
df_ds1 = pd.read_csv(f"{path}/SOCR-HeightWeight.csv")

# Extract Height column
data_ds1 = df_ds1['Height(Inches)'].values

# Create boxplot for Dataset 1
sns.boxplot(data=data_ds1)
plt.show()
```

# Task 5: Analysis of Dataset 1 - Height Distribution

The boxplot shows a **symmetric distribution** with the median line centered in the box. Both whiskers extend equally in length, and outliers appear on both ends in balanced numbers. This pattern is characteristic of a **normal distribution**.

---
## Dataset 2: Customer Satisfaction (Normal Distribution)

```python
# Download and load Dataset 2: Apple iPhone Sales
path = kagglehub.dataset_download("vengateshvengat/apple-iphone-complete-data-2007-2025")
df_ds2 = pd.read_csv(f"{path}/iphone_india_full.csv")

# Extract customer_satisfaction_percent column
data_ds2 = df_ds2['customer_satisfaction_percent'].values

# Create boxplot for Dataset 2
sns.boxplot(data=data_ds2)
plt.show()
```

# Task 5: Analysis of Dataset 2 - Customer Satisfaction Distribution

The boxplot displays a **relatively symmetric distribution** with the median centered in the box. Both whiskers extend fairly equally, and outliers appear on both ends. Despite some dense outlier clusters, the overall symmetric pattern confirms this is a **near-normal distribution**.

---
## Dataset 3: Ads Per Week (Non-Normal Distribution)

```python
# Download and load Dataset 3: Spotify Churn Analysis
path = kagglehub.dataset_download("nabihazahid/spotify-dataset-for-churn-analysis")
df_ds3 = pd.read_csv(f"{path}/spotify_churn_dataset.csv")

# Extract ads_listened_per_week column
data_ds3 = df_ds3['ads_listened_per_week'].values

# Create boxplot for Dataset 3
sns.boxplot(data=data_ds3)
plt.show()
```

# Task 5: Analysis of Dataset 3 - Ads Per Week Distribution (Non-Normal)

This boxplot shows a **right-skewed distribution**. The box is compressed toward the lower end, with almost no lower whisker. The upper whisker extends much farther, and many outliers cluster on the upper side. This asymmetric pattern indicates a **non-normal distribution**.

# Task 6: How Can You Test Your Data for Being Normally Distributed?

We can test for normal distribution by creating a boxplot and examining its visual characteristics.

**Visual Indicators of Normal Distribution:**
- Symmetric box with centered median line
- Equal whisker lengths on both sides
- Outliers balanced on both ends

**Visual Indicators of Non-Normal Distribution:**
- Asymmetric box shifted to one side
- Unequal whisker lengths
- Outliers concentrated on one end only

Boxplots provide quick visual assessment of normality without requiring assumptions about the underlying distribution.
